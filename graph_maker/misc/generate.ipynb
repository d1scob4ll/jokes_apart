```python
model_name = "gemma3:4b" 
num_items = 20
jokes_filename = "jokes_generated.txt"
sentences_filename = "sentences_generated.txt"
```

```python
joke_prompt_template = "Напиши одну законченную смешную шутку на русском языке на тему: {тема}. Не оставляй эмодзи."
sentence_prompt_template = "Напиши одно вырванное из контекста предложение содержащее от 8 до 20 слов на русском языке на тему: {тема}. Без кавычек." 

def generate_text(prompt, model):
    try:
        response = ollama.generate(model=model, prompt=prompt)
        return response.get('response', '').strip()
    except Exception as e:
        print(f"Ошибка при генерации текста: {e}")
        return ""

def generate_jokes(themes, model, num_items, filename, prompt_template):
    print(f"Начинаем генерацию {num_items} шуток для каждой темы...")
    with open(filename, "w", encoding="utf-8") as f:
        for theme in themes:
            print(f"  Тема: {theme}")
            for i in tqdm.tqdm(range(num_items), desc=f"    Генерация шуток для '{theme}'"):
                prompt = prompt_template.format(тема=theme)
                joke = generate_text(prompt, model)
                if joke:
                    f.write(f"\n{joke}\n")
                else:
                    f.write(f"Тема: {theme}, Шутка #{i+1}: <Не удалось сгенерировать>\n")
    print(f"Генерация шуток завершена. Результат сохранен в {filename}")

def generate_sentences(themes, model, num_items, filename, prompt_template):
    print(f"Начинаем генерацию {num_items} предложений для каждой темы...")
    with open(filename, "w", encoding="utf-8") as f:
        for theme in themes:
            print(f"  Тема: {theme}")
            for i in tqdm.tqdm(range(num_items), desc=f"    Генерация предложений для '{theme}'"):
                prompt = prompt_template.format(тема=theme)
                sentence = generate_text(prompt, model)
                if sentence:
                    f.write(f"\n{sentence}\n")
                else:
                    f.write(f"Тема: {theme}, Предложение #{i+1}: <Не удалось сгенерировать>\n")
    print(f"Генерация предложений завершена. Результат сохранен в {filename}")

if __name__ == "__main__":
    generate_jokes(themes, model_name, num_items, jokes_filename, joke_prompt_template)

    generate_sentences(themes, model_name, num_items, sentences_filename, sentence_prompt_template)

    print("Скрипт завершен.")
```
